# Tutorial on mcerp

https://pythonhosted.org/mcerp/

UV : uncertain Variable

liste of distributions : https://pythonhosted.org/mcerp/distribution_constructors.html

```python
import mcerp
mcerp.npts = 12_000
```

```python
x = mcerp.Normal(1, 3, tag="toto")
y = mcerp.Normal(2, 5)
print(x)
print(y)
print(x.mean)
print(x.var)
print(x.kurt)
print(x.skew)
print(mcerp.npts)
print(x.tag)
```

```python
x.describe()
y.describe()
x.plot()
y.plot()
print(x.tag)
```

```python
z = x + y
```

```python
z.describe()
z.plot()
```
